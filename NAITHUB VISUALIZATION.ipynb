```python
# from a pandas dataframe
batsman = pd.read_csv('Session_23_on_BasicsOfMatplotlibc-main/NAIT DS.csv')
batsman




plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(
    batsman['index'], 
    batsman['V Kohli'], 
    marker='o', linestyle='-', color='blue', linewidth=2, markersize=8, label='V Kohli Runs'
)

# Enhancements
plt.xlabel('Match Number', fontsize=12, fontweight='bold', color='darkblue')
plt.ylabel('Runs Scored', fontsize=12, fontweight='bold', color='darkred')
plt.title('Virat Kohli Performance Over Matches', fontsize=14, fontweight='bold', color='darkgreen')

plt.grid(True, linestyle='--', alpha=0.7)  # Add a transparent grid
plt.xticks(fontsize=10, fontweight='bold', color='black')
plt.yticks(fontsize=10, fontweight='bold', color='black')
plt.legend()  # Show legend

# Show the plot
plt.show()
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```
